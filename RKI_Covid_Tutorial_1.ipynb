```python
# Die Sprache python ist für sich allein schon recht flexibel, aber die großen Möglichkeiten kommen erst, wenn man
# externe Module zum Basisprogramm dazulädt.
# Wir benötigen zwei Module, die mit großen Datenreihen und Datenfeldern umgehen können. Dies sind zum einen numpy,
# ein Modul, was mit Zahlen und matrizen rechnen kann und
# pandas, eine Art Datenmanager für Tabellen, mit enormen Sortier, Ordnungs und Filterfuntionen.

# Das importieren der Module geht einfach in der Form

import numpy as np
import pandas as pd

# Jetzt können wir alle Zusatzbefehle und Strukturen von Pandas nutzen unter den Namen np und pd
```

```python
# Unser erstes Ziel ist es, die Covid-daten zu laden.
# Zur Sicherheit wechseln wir in unserem Jupyter Notebook erst einmal in das 
# richtige Arbeitsverzeichnis, dorthin, wo unsere Covid Datenbank liegt.

%cd ~/Documents/Work/Tutorials/RKI_Covid

# Wir definieren nun eine Variable, die den Namen unserer Datenbank enthält. 
# Da wir uns ja schon im richtigen Arbeitsverzeichnis befinden, 
# brauchen wir hier den Pfad nicht mehr mit eingeben 
#covid_path = "211203_RKI_COVID19.csv"

covid_path = "211203_RKI_COVID19_sample.csv"


#Achtung: Im Github ist nur ein kleiner Auszug der Original RKI Covid_Datei vom 03.12.21 abgelegt.
# Hinweis zu den Daten (Stand 03.12.2021):
# Die Daten sind die „Fallzahlen in Deutschland“ des Robert Koch-Institut (RKI) und stehen unter der Open Data Datenlizenz Deutschland – Namensnennung – Version 2.0 zur Verfügung. 
# Quellenvermerk: Robert Koch-Institut (RKI), dl-de/by-2-0
```

```python
# Jetzt kommt der erste wirklich mächtige Befehl aus der Pandas Bibliothek, nämlich
# der Befehl "read_csv", der unsere Datenbank, die ja im csv Format vorliegt einlädt.
# Da der Befehl ja aus einer externen Bibliothek kommt, müssen wir das Bibliothekskürzel pd voranstellen
# Der Befehl liefert als Ergebnis eine Pandas Datenstruktur, einen sogenannten Dataframe
# den wir der Variable covid zuweisen. Der Dataframe oder übersetzt datenrahmen ist eine der beiden mächtigen Datenstrukturen von Pandas, 
# dazu aber mehr in einem anderen Tutorial.

covid = pd.read_csv(covid_path)
```

```python
# Schauen wir uns den Inhalt der des dataframes covid nun einmal an. Dazu rufen wir covid einfach auf.

covid
```

```python
# Das ist doch schon einmal sehr beeindruckend. Wir sehen hier die Übersicht einer 
# Datentabelle mit über 2 Millionen Zeilen und 18 Spalten. Und der Begriff 
# Dataframe oder Datenrahmen kommt hier bildlich sogar irgendwie gut rüber.

# Ganz links am Rand sehen wir eine Durchnummerierung der ganzen Zeilen, den sog. Index
# In der Mitte sehen wir anhand der vielen Dreipunktsequenzen, dass die Ausgabe unterbrochen ist,
# es sollen ja nicht alle 2 Millionen Zeilen vor unserem Auge durchscrollen.

# Wir können uns auch schon einmal die Spalten unserer Datenbank ansehen, und was da so drin ist.
# Die Spaltennamen sprechen zum Teil für sich und wir picken uns hier schon einmal ein paar wichtige Infos raus:

# Da gibt es z.B. das Bundesland, den Landkreis, die Altersgruppe und das Geschlecht, aber auch Das Meldedatum

# Und um die Sache kurz zu machen und ohne zu spoilern, diese datenbank enthält die einzelnen 
# Meldungen aus den Landkreisen für ein bestimmtes Datum. Nicht jeder Landkreis meldet jeden Tag, aber in Summe
# erhalten wir hier einen Übeblick über jeden Tag der Pandemie.
```

```python
# mit dem dataframe covid können wir nun arbeiten. Wir können diverse Funktionen auf ihn anwenden.
# So können wir zum Beispiel mit der Funktion info(), eine Kurzzusammenfassung zu unseren Daten erzeugen.
# Pythom ist objetorientiert, deswegen wenden wir die Funtion info auf unseren Dataframe covid 
# über die Punkt Schreibweise an:

covid.info()
```

```python
# Hier werden sie 18 Spalten aufgelistet und die Datentypen, die dort vorhanden sind, d.h. ob 
# eine Spalte zum Beispiel reine Zahlenwerte enthält. Darum kümmern wir uns aber später.
```

```python
# ich will nun noch ein paar erste Möglichkeiten zeigen, wie man auf so einen Dataframe zugreifen kann:

# zum Beispiel kann ich ganz einfach eine oder mehrere Spalten rausgreifen:
    
covid['Bundesland']    
```

```python
covid[['Meldedatum', 'Bundesland']]
```

```python
# Was fällt uns denn hier auf? Die Darstellung ist unterschiedlich, 
# ob wir eine oder mehrere Spalten anzeigen lassen.

# Wenn wir zwei Spalten rausgreifen, bekommen wir als Ergebnis wieder
# einen dataframe, nur das er diesmal nur zwei Spalten und den Index hat

# Beim Rausschneiden einer Spalte bekommen wir es mit der zweiten Datenstruktur von
# Pandas zu tun, der so. Series.
# Eine Series ist also wie ein Spaltenvektor, ein Dataframe ist aus mehreren Series zusammengesetzt.
# Beide haben eine Index, den man jeweils als ganz linke Spalte laufen sieht.
```

```python
# Bevor es nun zu viel Stoff wird noch zwei kleinere Befehle, mit denen man weiter analysieren kann:
# Die Funktione value_counts zeigt uns, wie viele verschiedene Elemente in einer Spalte sind, und wie oft 
# die jeweils in der Spalte auftauchen

# Wenden wir doch einfach mal die Funktion value_counts() auf unsere Series an, die nur das Bundesland enthält:

covid['Bundesland'].value_counts()
```

```python
# Das Gleiche können wir auch mit der Landkreis spalte machen:


covid['Landkreis'].value_counts()


# Hier wird die Ausgabe wieder kompat gekürzt, denn es sind ja 313 verschiedene Landkreistypen 
# dort in der Spalte vorhanden und die sollen ja nicht einfach vor unserem auge durchscollen.
```

```python
# Zum Abschluss wollen wir noch die erste Rechenoperation auf unsere datenbank anwenden, oder
# besser gesagt, auf eine Spalte der datenbank

# Da jede Zeile ja die Fallmeldung enthält wollen wir diese mal aufsummieren um die Gesamtzahl 
# der gemeldeten Infektionen zu bestimmen

# Dazu greifen wir uns wieder eine Spalte rasu, in dem Fall die Spalte mit den Fallzahlen 
# und wenden die Funktion sum(), also Summenbildung auf die Spalte an.

covid['AnzahlFall'].sum()

#Anmerkung: Da wir hier nur einen Ausschnitt der Original-Datei untersuchen ist die Fallzahl geringer.
```
